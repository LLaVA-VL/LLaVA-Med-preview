```python
import numpy as np  
import matplotlib.pyplot as plt  
fig = plt.figure(figsize=(6,3)) # Create matplotlib figure

instruct_data = (102 +  90 + 50 + 158)/1000.0

# Create sample data  
x = np.array([1, 2, 3, 4, 5])  

y1 = np.array([0, 0, 0.158, 0.0035, instruct_data])  
y2 = np.array([129, 15, 0.595, 5, 1112])  
model_names = ['BLIP2', 'OpenFlamingo', 'LLaVA', 'MiniGPT4', 'mPLUG-Owl']

for xx, yy1 in zip(x, y1):
    if yy1 != 0:
        plt.text(xx-.3, 0.005, f'{yy1*1000}K', color='k', fontsize=10)


# Add text on top of each bar  
for i, v in enumerate(y2):  
    plt.text(i+0.7, v+1, str(v)+'M', color='k', fontsize=10)  

# Plot the first set of bars  
plt.bar(x, y1, color='orange', label='Instruction-following data')  
  
# Plot the second set of bars on top of the first set  
plt.bar(x, y2, color='skyblue', bottom=y1, tick_label=model_names, label='Image-text pairs')  
  
# Add labels and title  
plt.ylabel('# Training samples (Millions)')  
plt.yscale('log') 
  
plt.legend() 
# Show the plot  
plt.show()  

fig.savefig('output/barchart_training_data.pdf', bbox_inches='tight')
```
